```python
iter_ = next(iter(traindata))
```

```python
_a = 0
for i in iter_[0][0][0]:
    _a +=1
    if _a <= 10:
        print(i)
```

```python
index_total = []
len = []

for i in traindata:
    for k, ind in enumerate(i[-1]):
        len.append(i[1][0][k])
        index_total.append(ind)
```

```python
print(iter_[1][0])
print(iter_[-1])
```

```python
for ind, index in enumerate(index_total):
    print(len[ind], text_file[index])
```

```python
length = []

for i in text_file.keys():
    print(len(text_file[i]), text_file[i])
    length.append(len(text_file[i]))
```
